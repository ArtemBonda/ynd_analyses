```python
def get_counts_event_of_df(df, source):
    """Функция возвращает новую таблицу, в которой происходит соединение двух подсчетов количества событий в 
       наборе и доля событий к набору
       
    Args:
        df (pd.DataFrame): исходная таблица.
        source (str): колонка по которой происходит подсчет
    Returns:
        pd.DataFrame: таблица с новыми признаками.
    """
    temp1 = df[source].value_counts().rename('cnt')
    temp2 = df[source].value_counts(normalize=True).rename('% of value').apply(lambda x: x*100)
    event_counts = pd.concat([temp1, temp2], axis=1)

    return event_counts

# Вывод результата по количеству событий в наборе
get_counts_event_of_df(mobile_df, 'event.name')
```

В наборе нету событий "Установка" и 'Регистрация', чтобы можно было отследить первое появление пользователя в приложениию. Отберем из набора даты их первого появления в приложении и будем это считать что пользователь установил и зарегистрировался в приложении. Но в жизни бывает что пользоватеь скачал приложение, но даже его не запустил и тем более не зарегистриловался. Так что "Установка" != "Регистрация"

В начале подготовим функции для подсчетов и визуализации

```python
def get_profiles(df_sessions):
    
    """Функция создания пользовательских профилей пользователей на основании первого ресурса 
       привлечениия и первого появления пользователя

    Args:
        df_sessions (pd.DataFrame): исходная таблица содержащая сессии пользователей

    Returns:
        pd.DataFrame: таблица с датами регистрацией пользователя.
    """
    profiles = (
            df_sessions.sort_values(by=['user_id', 'event_time']) # Сортируем по uid, и  дате и времени начала сессии
            .groupby('user_id')                                   # сгруппировать пo uid
            .agg({'event_time': 'first', 'source': 'first'})      # Нацдем первый раз посещения с 'channel' которого зашли  
            .rename(columns={'event_time': 'first_ts'})           # Сразу переименум колонки
            .reset_index()                                        # возвращаем user_id из индекса
        )
    
    profiles['date'] = profiles['first_ts'].dt.date
    profiles['week'] = profiles['first_ts'].dt.isocalendar().week
    
    return profiles

# Подготовка функции к визуализации событий во времени
def get_bar_by(df, by_group, columns, aggfunc='count', rotation=0, title_name='событий'):
    
    """Функция выводящая на экран диаграму по срупированному полю, делает по умолчанию подсчет событий в columns

    Args:
        df (pd.DataFrame): исходная таблица.
        by_group (str): столбец по которому производится группировка в наборе
        columns (str): столбец по которому производится подсчет 
        aggfunc (): по умолчанию - "count", агрегирующая функция которая производит операции в columns
    Returns:
        px.bar(): столбчатая диаграмма
    """
    
    # готовим сводную таблицу
    #table = df.groupby(by_group).agg({columns:aggfunc})
    table = df.pivot_table(index=by_group, aggfunc={columns:aggfunc})
    # строим линейный график
    fig = px.bar(table, x=table.index, y=table[columns], text_auto=True)
    
    # оформляем график
    fig.update_layout(
        width= 1000, 
        height=400,
        font=dict(size=16),
        title=dict(
            text=f'<b>Распределение количества {title_name} по {by_group}</b>',
            font=dict(size=22),
            xanchor='center',
            yanchor='top',
            x = 0.5,
            y= 0.95,
        ),
        annotations=[
            dict(
                text=f'<b>Note<sup>1</sup></b> :<br> Количество произошедшее с {table.index[0]} по {table.index[-1] } {by_group}',
                xref='paper',
                yref='paper',
                x=0,
                y=-0.3,
                align='left',
                showarrow=False, 
                font=dict(size=13),
            )  
        ],
        showlegend=False,
        plot_bgcolor='#ffffff',
        xaxis_title=by_group,
        yaxis_title=f'Количество {title_name}',
        margin=dict(l=0, r=0))
    
     # вывод подсказки, оформление колонок  
    fig.update_traces(
        hoverinfo="all", hovertemplate="Количество: %{y}", 
        marker=dict(line=dict(color='#000000',width=.5)),
        opacity=0.9)
    
    fig.update_xaxes(tickangle=rotation, tickmode = 'linear')
    
    fig.show()
```

```python
#  Получаем набор уникольных пользователй и перове появление
profiles = get_profiles(df)
# соеденим с таблицей пользователей совершивших целевое событие,
# пропущенные значения заменим на 0 и False соотвественно
profiles = profiles.merge(looked_contacts, on='user_id', how='left').fillna({'cnt_look_contacts':0, 'is_show_contacts':False})
profiles['cnt_look_contacts'] = profiles['cnt_look_contacts'].astype(int)
```

### Выделение сессии

Выделим таблицу событий отсортированных по пользователю и времени события

```python
# Сортируем юзерам и событиям
sorted_df = df[['user_id','event_time', 'event_name']].copy().sort_values(['user_id', 'event_time']).reset_index()

# считаем разницу между временем посещения страницы и временем посещения другой страницы
# Если страница была первой для пользователя, то значение в колонке ’diff’ будет NaT, т. к. нет предыдущего значения
sorted_df['diff'] = sorted_df.groupby('user_id')['event_time'].diff(1)
sorted_df = sorted_df.rename(columns={'index':'id'})
```

Cоздадим вспомогательный датафрейм `session_start_df`. Этот датафрейм будет содержать события, которые будут считаться первыми событиями сессий. К таким событиям относятся все события, которые произошли спустя более чем 30 минут после предыдущего либо события, которые были первыми для пользователя (NaT в колонке `diff`). 30 минут это достаточно долгое время отсутсвия в приложении чтобы считать что пользователь, закрыл и не пользовался им планомерно, а не отвлекся на что то, что у нас обычно в жизни может занять 10 - 20 минут. 

Также создадим во вспомогательном датафрейме колонку `session_id`, которая будет содержать в себе `index` первого события сессии. Она пригодится, чтобы корректно отобразить идентификатор сессии, когда будем объединять данные из основного и вспомогательного датафреймов.

```python
# Создадим вспомогталельный набор с
session_start_df = sorted_df[(sorted_df['diff'].isnull()) | (sorted_df['diff'] > '1800 seconds')].copy()
session_start_df['session_id'] = session_start_df.loc[:,'id']
# session_start_df
```

```python
# С помощью функции merge_asof объединим между собой данные основного и вспомогательного датафреймов
sorted_df = sorted_df.sort_values('id')
session_start_df = session_start_df.sort_values('id')
sorted_df = pd.merge_asof(sorted_df, session_start_df[['id', 'user_id', 'session_id']], on='id', by='user_id')
```

```python
# Найдем события, которые были первыми в сессиях.
sorted_df['is_first_event_in_session'] = sorted_df['id'] == sorted_df['session_id']
```

```python
# Вычисляем время, разницу между предыдущей и следующей страницей внутри сессии
sorted_df['time_on_page_sec'] = sorted_df.groupby(['session_id'])['event_time'].diff(1)
```

```python
#  сдвинуть значение столбца `time_on_page` на одну строку вверх внутри отдельно взятой сессии
sorted_df['time_on_page_sec'] = sorted_df.groupby(['session_id'])['time_on_page_sec'].shift(-1)
```

```python
# преобразуем ’time_on_page’ в секунды, для удобства арифметических операций
sorted_df['time_on_page_sec'] = sorted_df['time_on_page_sec'] / np.timedelta64(1, 's')
```

### Разделение пользователей на группы

### Когортный анализ

Соеденим таблицу профилей пользователей с очищеной таблицей сессий. Профили соодержат о первых визитах юзеров в приложение, но для расчета Ratention Rate нужны данные обо всех посещениях из таблицы сессий  

```python
# доступный интервал привлечения пользователей
min_analysis_date = profiles['date'].min()

# задаём момент и горизонт анализа данных
observation_date = profiles.date.max()
# Задаем горизонт
horizon_days = 14
# создаём опцию «игнорировать горизонт»
ignore_horizon = False
```

```python
# тестовая
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],  
    ignore_horizon=False,
):
    
    '''
         Функция для расчёта коэффициента удержания пользователей

    Args:
        profiles (pd.DataFrame): исходная таблица содержащая профили пользователей
        sessionbs (pd.DataFrame): исходная таблица содержащая сессии пользователей
        observation_date (date): момент анализа
        horizon_days (int): горизонт анализа данных
        dimensions (): массив в котором передаются параметр(ы) для группировки таблицы удержания
        ignore_horizon (bool): опция 'игнорировать горизонт анализа', по умолчанию "False"
    Returns:
        result_raw (pd.DataFrame): сырые данные
        result_grouped (pd.DataFrame): таблица удержания
        result_in_time (pd.DataFrame) :таблицу динамики удержания.
    
    
    '''
    
    # добавляем столбец is_show_contacts в передаваемый dimensions список
    dimensions = ['is_show_contacts'] + dimensions
    
    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('date <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'event_time']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['event_time'] - result_raw['first_ts']
    ).dt.days
    
    def group_by_dimensions(df, dims, horizon_days):
    
        '''
             Функция для расчёта
        Args:
            profiles (pd.DataFrame): исходная таблица содержащая профили пользователей
            dims (pd.DataFrame): исходная таблица содержащая сессии пользователей
            horizon_days (int): горизонт анализа данных

        Returns:
            result (pd.DataFrame): сформированные данные

        '''
        # строим «треугольную» таблицу
        result = df.pivot_table(index=dims, columns='lifetime', values='user_id', aggfunc='nunique')
        
        # определяем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        
        # добавляем размеры когорт к «треугольной» таблице
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        
        # делим каждую «ячейку» на соответствующий размер когорты и получаем retention rate
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        return result
    
    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(result_raw, dimensions + ['date'], horizon_days)
    
    # возвращаем сырые данные, таблицу удержания и таблицу динамики удержания
    return result_raw, result_grouped, result_in_time
```

```python
def get_conversion(
    profiles,
    events, 
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):
    
    '''
         Функция для расчёта коэффициента конверсии пользователей

    Args:
        profiles (pd.DataFrame): исходная таблица содержащая профили пользователей
        sessionbs (pd.DataFrame): исходная таблица содержащая сессии пользователей
        observation_date (date): момент анализа
        horizon_days (int): горизонт анализа данных
        dimensions (): массив в котором передаются параметр(ы) для группировки таблицы удержания
        ignore_horizon (bool): опция 'игнорировать горизонт анализа', по умолчанию "False"
    Returns:
        result_raw (pd.DataFrame): сырые данные
        result_grouped (pd.DataFrame): таблица удержания
        result_in_time (pd.DataFrame) :таблицу динамики удержания.
    
    
    '''

    # Шаг 1. Получить пользовательские профили и данные о покупках
    # передаём их в качестве аргументов profiles и events

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('date <= @last_suitable_acquisition_date')

    # Шаг 2. Найти дату и время первого целевого действия contacts_show для каждого пользователя
    first_show = (
        events.query('event_name == "contacts_show"')
        .sort_values(by=['user_id', 'event_time'])
        .groupby('user_id')
        .agg({'event_time': 'first'})
        .reset_index()
    )

    # Шаг 3. Добавить данные о событии в профили
    result_raw = result_raw.merge(first_show[['user_id', 'event_time']], on='user_id', how='left')

    # Шаг 4. Рассчитать лайфтайм для каждого события
    result_raw['lifetime'] = (result_raw['event_time'] - result_raw['first_ts']).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users'
        dimensions = dimensions + ['cohort']
        
    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        # строим «треугольную» таблицу конверсии
        result = df.pivot_table(index=dims, columns='lifetime', values='user_id', aggfunc='nunique')
        
        # считаем сумму с накоплением для каждой строки
        result = result.fillna(0).cumsum(axis = 1)
        
        # вычисляем размеры когорт
        cohort_sizes = (
            df.groupby(dims)
              .agg({'user_id': 'nunique'})
              .rename(columns={'user_id': 'cohort_size'})
        )
        
        # добавляем размеры когорт в таблицу конверсии
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        
        # делим каждую «ячейку» в строке на размер когорты и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        # исключаем все лайфтаймы, превышающие горизонт анализа
        result = result[['cohort_size'] + list(range(horizon_days))]
        # восстанавливаем размеры когорт
        result['cohort_size'] = cohort_sizes
        return result    
    

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['date'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

```python
def plot_retention(retention, retention_history, horizon, window=7):
    
    '''
         Функция отрисовки графика удержания пользователей в приложении, по умолчанию выставляется окно в 7 дней для 
         корректировки на недельном промежутке график. Функция отрисовывает 4 графика по переданным таблицам удержания 
         и динамики удержания во времени горизонта

    Args:
        retention (pd.DataFrame): исходная таблица содержащая удержание пользователей 
        retention_history (pd.DataFrame): исходная таблица содержащая удержание пользователей во времени
        horizon (int): горизонт анализа данных
        window (int): целое число, передается вспромагательной функции, которая выставляет окно для фильтрации(сглаживания)
        отображения графика, по умолчанию фильтрация происходит по недельному окну
    Returns:
         (): график удержания по полю с флагом True
         (): график удержания по полю с флагом False
         (): график динамики удержания по полю с флагом True со сглаживанием данных
         (): график динамики удержания по полю с флагом False со сглаживанием данных
    
    '''

    # задаём размер сетки для графиков
    plt.figure(figsize=(15, 10))

    # исключаем размеры когорт и удержания первого дня
    retention = retention.drop(columns=['cohort_size', 0])

    # если в индексах таблицы удержания только is_show_contacts,
    # добавляем второй признак — cohort
    if retention.index.nlevels == 1:
        retention['cohort'] = 'All users'
        retention = retention.reset_index().set_index(['cohort', 'is_show_contacts'])

    # в таблице графиков — два столбца и две строки, четыре ячейки
    # в первой строим кривые удержания просматривающих контакты пользователей
    ax1 = plt.subplot(2, 2, 1)
    retention.query('is_show_contacts == True').droplevel('is_show_contacts').T.plot(
        grid=True, ax=ax1
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание не просматривающих контакты пользователей')

    # во второй ячейке строим кривые удержания не просматривающих контакты
    # вертикальная ось — от графика из первой ячейки
    ax2 = plt.subplot(2, 2, 2, sharey = ax1)
    retention.query('is_show_contacts == False').droplevel('is_show_contacts').T.plot(
        grid=True, ax=ax2
    )
    plt.legend()
    plt.xlabel('Лайфтайм')
    plt.title('Удержание просматривающих контакты пользователей')

    # в третьей ячейке — динамика удержания просматривающих контакты
    ax3 = plt.subplot(2, 2, 3)
    
    # получаем названия столбцов для сводной таблицы
    columns = [
        name
        for name in retention_history.index.names
        if name not in ['date', 'is_show_contacts']
    ]
    
    # функция filter_data() для сглаживания фрейма с недельным окном
    def filter_data(df, window):
        # для каждого столбца применяем скользящее среднее
        for column in df.columns.values:
            df[column] = df[column].rolling(window).mean() 
        return df


    # фильтруем данные, строим сводную таблицу и график по ней
    filtered_data = retention_history.query('is_show_contacts == True').pivot_table(
        index='date', columns=columns, values=horizon-1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax3)
    plt.xlabel('Дата привлечения')
    plt.xticks(rotation=-20, ha='left')
    plt.title(
        'Динамика удержания просматривающих контакты пользователей на {}-й день'.format(
            horizon
        )
    )
    
    # в четвёртой ячейке — динамика удержания несмотрящих контакты
    ax4 = plt.subplot(2, 2, 4, sharey = ax3)
    # фильтруем данные, строим сводную таблицу и график по ней
    filtered_data = retention_history.query('is_show_contacts == False').pivot_table(
        index='date', columns=columns, values=horizon - 1, aggfunc='mean'
    )
    filter_data(filtered_data, window).plot(grid=True, ax=ax4)
    plt.xlabel('Дата привлечения')
    plt.xticks(rotation=-20, ha='left')
    plt.title(
        'Динамика удержания не просматривающих контакты пользователей на {}-й день'.format(
            horizon
        )
    )

    # исключаем наложение графиков друг на друга
    plt.tight_layout()

    plt.show()
```

```python
def add_features(df):
    """Функция генерарирующая новую таблицу, в которой передается таблица с событиями. в которой 
    происходит группировка по user_id , event_time и создание записей переходов пользователей по событиям

    Args:
        df (pd.DataFrame): исходная таблица.
    Returns:
        pd.DataFrame: таблица с новыми признаками.
    """
    # Сортировка по user_id , event_time
    sorted_df = df.sort_values(by=['session_id', 'event_time']).copy()
    # добавить шаги событий
    sorted_df['step'] = sorted_df.groupby('user_id').cumcount() + 1
    
    # добавить узлы источники и узлы целей
    # узлы источники - это сами события
    sorted_df['nod_source'] = sorted_df['event_name']
    # добавить целевые узлы
   
    sorted_df['nod_target'] = sorted_df.groupby('user_id')['nod_source'].shift(-1)
    # Вернем таблицу без источника регистрации и без имени событий
    return sorted_df.drop(columns=['event_name'])
```

Создание словаря индексов для `nod_source`, где на каждом следующем шаге `nod_target` становится `source`, чтобы диаграмма гененрировалась корректно нужна правильная индексация nod_source на каждом шаге.

```python
def get_dict_source_idx(df):
    """Функция генерации индексов source

    Args:
        df (pd.DataFrame): исходная таблица с признаками step, nod_source, nod_target.
    Returns:
        dict: словарь с индексами, именами и соответсвиями индексов именам source.
    """
    # инициализируем словарь
    res_dict = {}
    
    count = 0
    
    # Получить индексы источников
    for nod, step in enumerate(df['step'].unique().tolist()):
        # получить уникальное наименование для шага
        res_dict[nod+1] = {}
        res_dict[nod+1]['sources'] = df[df['step'] == step]['nod_source'].unique().tolist()
        res_dict[nod+1]['sources_index'] = []
        for i in range(len(res_dict[nod+1]['sources'])):
            res_dict[nod+1]['sources_index'].append(count)
            count += 1
    
    # Соеденнение списков
    for key in res_dict:
        res_dict[key]['sources_dict'] = {}
        for name, nod in zip(res_dict[key]['sources'], res_dict[key]['sources_index']):
            res_dict[key]['sources_dict'][name] = nod
    # вернуть словарь 
    return res_dict
```

```python
def colors_for_sources(mode):
    
    """Генерация цветов rgba

    Args
        mode (str): сгенерировать случайные цвета, если 'random', а если 'custom' - 
                    использовать заранее подготовленные
    Returns:
        dict: словарь с цветами, соответствующими каждому индексу
    """
    # словарь, в который сложим цвета в соответствии с индексом
    colors_dict = {}
    
    if mode == 'random':
        # генерим случайные цвета
        for label in df_comp['nod_source'].unique():
            r, g, b = np.random.randint(255, size=3)            
            colors_dict[label] = f'rgba({r}, {g}, {b}, 1)'
            
    elif mode == 'custom':
        # присваиваем ранее подготовленные цвета
        colors = requests.get('https://raw.githubusercontent.com/rusantsovsv/senkey_tutorial/main/json/colors_senkey.json').json()
        for no, label in enumerate(df_comp['nod_source'].unique()):
            colors_dict[label] = colors['custom_colors'][no]
            
    return colors_dict
```

```python
def percent_users(sources, targets, values):
    
    """
    Расчет уникальных id в процентах (для вывода в hover text каждого узла)
    
    Args:
        sources (list): список с индексами source.
        targets (list): список с индексами target.
        values (list): список с "объемами" потоков.
        
    Returns:
        list: список с "объемами" потоков в процентах
    """
    
    # объединим источники и метки и найдем пары
    zip_lists = list(zip(sources, targets, values))
    
    new_list = []
    
    # подготовим список словарь с общим объемом трафика в узлах
    unique_dict = {}
    
    # проходим по каждому узлу
    for source, target, value in zip_lists:
        if source not in unique_dict:
            # находим все источники и считаем общий трафик
            unique_dict[source] = 0
            for sr, tg, vl in zip_lists:
                if sr == source:
                    unique_dict[source] += vl
                    
    # считаем проценты
    for source, target, value in zip_lists:
        new_list.append(round(100 * value / unique_dict[source], 1))
    
    return new_list
```

```python
def lists_for_plot(source_indexes, colors, frac=10):
    
    """
    Создаем необходимые для отрисовки диаграммы переменные списков и возвращаем
    их в виде словаря
    
    Args:
        source_indexes (dict): словарь с именами и индексами source.
        colors (dict): словарь с цветами source.
        frac (int): ограничение на минимальный "объем" между узлами.
        
    Returns:
        dict: словарь со списками, необходимыми для диаграммы.
    """
    
    sources = []
    targets = []
    values = []
    labels = []
    link_color = []
    link_text = []

    # проходим по каждому шагу
    for step in sorted(df_comp['step'].unique()):
        if step + 1 not in source_indexes:
            continue

        # получаем индекс источника
        temp_dict_source = source_indexes[step]['sources_dict']

        # получаем индексы цели
        temp_dict_target = source_indexes[step+1]['sources_dict']

        # проходим по каждой возможной паре, считаем количество таких пар
        for source, index_source in temp_dict_source.items():
            for target, index_target in temp_dict_target.items():
                # делаем срез данных и считаем количество id            
                temp_df = df_comp[(df_comp['step'] == step)&(df_comp['nod_source'] == source)&(df_comp['nod_target'] == target)]
                value = len(temp_df)
                # проверяем минимальный объем потока и добавляем нужные данные
                if value > frac:
                    sources.append(index_source)
                    targets.append(index_target)
                    values.append(value)
                    # делаем поток прозрачным для лучшего отображения
                    link_color.append(colors[source].replace(', 1)', ', 0.2)'))
                    
    labels = []
    colors_labels = []
    for key in source_indexes:
        for name in source_indexes[key]['sources']:
            labels.append(name)
            colors_labels.append(colors[name])
            
    # посчитаем проценты всех потоков
    perc_values = percent_users(sources, targets, values)
    
    # добавим значения процентов для howertext
    link_text = []
    for perc in perc_values:
        link_text.append(f"{perc}%")
    
    # возвратим словарь с вложенными списками
    return {'nod_sources': sources, 
            'nod_targets': targets, 
            'values': values, 
            'labels': labels, 
            'colors_labels': colors_labels, 
            'link_color': link_color, 
            'link_text': link_text}
```

```python
def plot_senkey_diagram(data_dict):    
    
    """
    Функция для генерации объекта диаграммы Сенкей 
    
    Args:
        data_dict (dict): словарь со списками данных для построения.
        
    Returns:
        plotly.graph_objs._figure.Figure: объект изображения.
    """
    
    fig = go.Figure(data=[go.Sankey(
        domain = dict(
          x =  [0,1],
          y =  [0,1]
        ),
        orientation = "h",
        valueformat = ".0f",
        node = dict(
          pad = 50,
          thickness = 15,
          line = dict(color = "black", width = 0.1),
          label = data_dict['labels'],
          color = data_dict['colors_labels']
        ),
        link = dict(
          source = data_dict['nod_sources'],
          target = data_dict['nod_targets'],
          value = data_dict['values'],
          label = data_dict['link_text'],
          color = data_dict['link_color']
      ))])
    
    fig.update_layout(
    hovermode = 'x',
        title="Диаграмма переходов пользователей в приложении",
        width=1000, height=1200,
        font=dict(size = 12, color = 'white'),
        plot_bgcolor='black',
        paper_bgcolor='black'
)
    
    # возвращаем объект диаграммы
    return fig
```

```python
def create_funnel_df(df, steps, from_date=None, to_date=None, step_interval=0):
    # фильтровать df только для событий в списке шагов
    """
    Функция принимает набор данных событий, список последовательных шагов и возвращает количество пользователей на этом шаге
    df (pd.Dataframe): исходный набор событий
    steps ([]): последовательность событий по шагам
    from_date(datatime): фильтр, дата с которой необходимо произвести расчеты, по умолчанию дата не задана
    to_date (datatime): фильтр, дата по которую необходимо произвести расчеты
    step_interval (int): 
    Args:
        data_dict (dict): словарь со списками данных для построения.
        
    Returns:
        plotly.graph_objs._figure.Figure: объект изображения.
    """
    df = df[['user_id', 'event_name', 'event_time']]
    df = df[df['event_name'].isin(steps)] # доп. проверка что бы были только эти события
#     events_for_funnel
    dfs = {}
    values = []
    
    # для каждого шага в воронке создаю df и отфильтрую только для этого шага
    for idx, step in enumerate(steps):
        if idx == 0:
            # создать dict для хранения отфильтрованного фрейма данных каждого шага
            
            
            # Сортировка по юзерам, которые совершили 1-оесобытие и найти минимальное время
            dfs[step] = df.query('event_name == @step')\
                            .sort_values(['user_id', 'event_time'], ascending=True) \
                            .drop_duplicates(subset=['user_id', 'event_name'], keep='first')
            
            # фильтр по дате
            if from_date:
                dfs[step] = dfs[step][(dfs[step]['event_time'] >= from_date)]
                
            if to_date:
                dfs[step] = dfs[step][(dfs[step]['event_time'] <= to_date)]
                                      
        else:
            # Иначе фильтровать по конкретным событиям
            dfs[step] = df[df['event_name'] == step]
            
            # Соединение с предыдущим шагом по user_id, что гарантирует попадание строк с предыдущего шага
            merged = pd.merge(dfs[steps[idx - 1]], dfs[step], on='user_id', how='outer')
            
            # keep only rows for which the distinct_ids appear in the previous step
            valid_ids = dfs[steps[idx - 1]]['user_id'].unique()
            merged = merged[merged['user_id'].isin(valid_ids)]
            
            # Сохранить только события произошедшие полсле предыдущего шага, с сортировкой по времени
            merged = merged[merged['event_time_y'] >= merged['event_time_x']].sort_values('event_time_y', ascending=True)
            
            # Выбрать минимльное время из допустимых для каждого пользователя
            merged = merged.drop_duplicates(subset=['user_id', 'event_name_x', 'event_name_y'], keep='first')
            
            # Выделим только необходимые колонки и переменуем их
            merged = merged[['user_id', 'event_name_y', 'event_time_y']].rename({'event_name_y':'event_name',
                                                                                  'event_time_y':'event_time'}, axis=1)
            # Встроим данные в dict чтобы можно быдо присоеденить к данным следующего шага
            dfs[step] = merged
            
        # добавить количество пользователей в список
        values.append(len(dfs[step]))
    
    # Создаем фрейм с получившимися данными
    funnel_df = pd.DataFrame({'event':steps, 'cnt_users':values})
    # Проценты для каждого шага
    funnel_df['pct'] = (100 -100 * abs(funnel_df['cnt_users'].pct_change()).fillna(0).astype(int))
    # Количество пользователей на прошлом шаге
    funnel_df['cnt_users-1_step'] = funnel_df['cnt_users'].shift(1)
    # Конверсия пользователей от первого шага
    funnel_df['conversion'] = (funnel_df['cnt_users'] / funnel_df['cnt_users'].loc[0] * 100).fillna(0).astype(int)
    
    return funnel_df[['event', 'cnt_users', 'conversion']]
```

Заполним таблицу событий номерами тестовых групп
- 1 - группа просмотревшая только tips_show
- 2 - группа просмотревшая tips_show и tips_click
- 0 - группа которая не попала под 2 первых условия 

```python
def category_test_group(row):
    
    '''
    Функия возвращает группу пользователей для проведения теста
    - 1, если  группа просмотревшая только tips_show
    - 2, если группа просмотревшая tips_show и tips_click
    - 0, группа которая не попала под 2 первых условия
    '''
    if row in  list_users_by_tips_show:
        return 1
    elif row in list_users_by_tips_show_and_tips:
        return 2
    else:
        return 0
```

### Формулирование гипотез

Сформулируем на собранных данных 2 гипотезы:

1. Одни пользователи совершают действия `tips_show` и `tips_click`, другие — только `tips_show`
2. Одни пользователи совершают действия только `tips_show`  и не совершают событие  `tips_click`,  другие — ни разу не увидят `tips_show`
3. Одни пользователи совершают действия `tips_show` и `tips_click`, другие — ни разу не увидят tips_show


#### Проверка гипотезы:
 **Гипотеза о равенстве пропорций просмотров контактов в  группах №1 И №2.**
- Одни пользователи совершают действия `tips_show` и `tips_click`, другие — только `tips_show`

Участники теста:
- `Группа #1` -  группа просмотревшая только tips_show
- `Группа #2` -  группа просмотревшая tips_show и tips_click

> Н0: конверсия в `просмотры контактов` равны у этих двух групп

> H1: конверсия в `просмотры контактов` различается у этих двух групп <- Гипотеза заказчика


Сравним доли пользователей совершивших действия `tips_show` и `tips_click` перешедщих к событию `contacts_show`. Какой можно сделать вывод о разнице в конверсии из Целевого действия в просмотр контактов для этих групп?

За критический уровень статической значимости возмем общеустоявшееся значение = 5% 

```python
# Функция подсчета А/В теста
def t_testing_ab(df, num_drop_group):
    """
    Функция вспомогательная для проведения А/В теста. Резултатом будет вывод на экран резултат проведения теста
    
    Args:
        df (pd.Dataframe):    исходная таблица с данными для проведения А/В теста
        num_drop_group (int): номер группы которой нужно удалить из теста 
        
    """
    # Установим cтатическую значимость
    alpha = 0.05
    # Возмем данные только групп 0 и 1, количество пользователей в событиях
    ab_test = df[['count_users', 'count_users_contacts_show']].drop(num_drop_group)
    
    # сохраним пользователей просмотревших контакты
    succes_show = np.array(ab_test['count_users_contacts_show'], ab_test['count_users_contacts_show'].values[1])
    # print(succes_show)
    
    # сохраним исходное количество пользователей в группах
    total_users = np.array(ab_test['count_users'], ab_test['count_users'].values[0])
    # print(total_users)
    
    p1 = succes_show[0]/total_users[0]
    p2 = succes_show[1]/total_users[1]
    difference = p1 - p2
    
    p_combined = (succes_show[0] + succes_show[1]) / (total_users[0] + total_users[1])
    
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/total_users[0] + 1/total_users[1]))
    
    distr = st.norm(0, 1)
    
    p_value = (1 - distr.cdf(abs(z_value))) * 2
    
    print('p-valeu: ', p_value)
    
    if p_value < alpha:
        print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
    else:
        print(
            'Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными'
        )
```

```python
#  А/В тест для групп 1 и 2, группа 0 удалить из подсчетов
t_testing_ab(df_hipotesis, 0)
```

Проверка гипотезы о равенстве долей подтвердило что, разница межу конверсиями в 17% и 30% является значимой. Стоит отметить что значение p_value это цифра которая находится на на 9 месте после запятой, это очень низкий показатель в %.
